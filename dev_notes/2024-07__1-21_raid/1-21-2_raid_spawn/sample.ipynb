```python
import math
import matplotlib.pyplot as mplt
import numpy
```

```python
STEP = math.pi * 0.001
RADIUS = 32 * 0.24
GRID_TEMPLATE = numpy.zeros((20, 20))
CENTER = (10, 10)
```

```python
count = math.floor(math.pi * 2 / STEP) + 1
grid = numpy.copy(GRID_TEMPLATE)

for i in range(count):
    angle = i * STEP
    x, y = RADIUS * math.cos(angle), RADIUS * math.sin(angle)
    grid[math.floor(x) + CENTER[0], math.floor(y) + CENTER[1]] += 1
```

```python
s = numpy.sum(numpy.sum(grid))
print(grid / s)
```

```python
mplt.imshow(grid, cmap='hot', interpolation='nearest')
mplt.colorbar()
mplt.show()
```

```python
combine_shape = (grid.shape[0], grid.shape[1])
combine = numpy.zeros(combine_shape)
for i in range(combine_shape[0]):
    for j in range(combine_shape[1]):
        end_coord_x, end_coord_y = min(i + 3, grid.shape[0]), min(j + 3, grid.shape[1])
        combine[i, j] = numpy.sum(numpy.sum(grid[i:end_coord_x, j:end_coord_y]))

mplt.imshow(combine, cmap='hot', interpolation='nearest')
mplt.colorbar()
mplt.show()
```
